* Sporadic missing data: Imputation을 통해서 데이터를 채우는 경우가 많음
* Longitudinal droupout: 시간에 따라 한 변량이 빠지면 그 이후에 채워지지 않음

`-` Missing data in Clinical trials

* 임상실험 시 보통 안좋은 결과에 의해서 Missing이 발생하는 경우가 많음 -> 단순히 빼고 분석하면 부적절

`-` Missing data

* MCAR
* MAR: 알고 있는 정보가 Missing과 연관되어 있음 -> 무시할 수 있는 Missing.
* MNAR: Missing 발생 사유 자체가 해당 변수의 값 자체에 연관되어 있음

`-` MNAR example - schizophrenia trial

* 기존 약품, 플라시보, 신약 그룹에 대해서 플라시보 그룹은 효과가 없기 때문에 시간이 지남에 따른 Missing이 더 많이 발생
* 약효에 대해서 Flat한 결과가 나와야 하는데, 효과가 없거나 악화되는 플라시보 그룹이 이탈하기 때문에 퍼포먼스 상승 곡선이 생겨남

`-` Missing data approaches

* Imputation approaches
* Inverse Probability weighting: Missing이 많이 발생한 그룹에 가중치를 크게 부여
* Joing Likelihood approaches: 가정을 통해 모델링
