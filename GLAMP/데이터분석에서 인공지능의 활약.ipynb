## 2. 딥러닝을 활용한 이상치 탐색

`-` Memorization effect

* 모형이 매우 유연하여 데이터를 완벽하게 분류할 수 있음. Easily Overfit
* 딥러닝이 데이터를 오버피팅할 때, 학습이 진행되면서 정상데이터를 먼저 오버피팅하고, 이상치를 나중에 맞추는 현상이 있음: ME
> Label Noise 문제에서 활용됨

`-` Label Noise Problem

* 몇몇 관측치에서 관측라벨이 실제 라벨과 다른 경우
* 실제 라벨을 알지 못하기 때문에 어려움 -> 사람의 실수, 인공지능 알고리즘의 부정확함에 따라 발생
* Label Noise 알고리즘을 활용하여 데이터 정련

`-` ME를 활용한 Label Noise 탐지 방법

* DNN을 데이터에 약하게 적합하고, bias가 큰 데이터포인트를 Label Noise로 선정

`-` ME를 활용한 이상치 탐색

* 일반적으로는 Robust하게 분포 추정 후 통계량을 활용 -> 기술적으로 어려움
* Deep Generate Model을 활용하여 분포 추정. VAE, GAN, Diffusion 등을 활용
> 생성 모델이 잘 생성하지 못하는 데이터를 이상치로 판단. ODIM

## 4. 예측모형 구축을 위한 Foundation Model

* In-context learning: 이전에 입력받은 텍스트를 활용해서 결과 개선
> 어떠한 파라미터 업데이트도 없음, LLM은 이미 세상의 모든 문서를 학습해 놓았고, 이걸 잘 뽑아내는 것이 Prompt engineering

* Input data $X$를 잘 제어하는 것이 프롬프트 엔지니어링

`-` 예측모형에서의 In-context learning

* 주어진 학습데이터를 prompt라 해석하면, LLM이 따로 파라미터의 학습 없이 예측이 가능할 수도 있을 것

`-` TabPFN

* SCM probability model가상의 데이터를 다양하게, 실제 상황과 비슷하게 생성하는 것
* 900만개 이상의 가상 데이터를 생성하여 LLM을 학습
* 퍼포먼스 뛰어남 ㅇㅇ
* TabPFN의 학습 데이터셋 규모를 매우 크게 확장하면 Categorical input도 수용 가능하고, regression/classification/robust regression/density estimation 모두 가능
* Missing data로 확장도 가능
